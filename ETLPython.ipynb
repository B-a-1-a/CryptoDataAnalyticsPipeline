```python
### Dependencies
import pandas as pd
import requests # API requests
import getpass
```

### Extract 

```python
url = "https://api.coincap.io/v2/assets"
```

```python
response = requests.get(url)
print(response)
# 200 means successful requests
```

```python
responsedata = response.json()
#responsedata
```

### Transform

```python
# Json returns "Semi-Structured" data
# We want structured data

data = pd.json_normalize(responsedata, "data")
data
```

```python
data.info()
```

```python
data["maxSupply"] = data["maxSupply"].fillna(0)
data["vwap24Hr"] = data["vwap24Hr"].fillna(0)
data["explorer"] = data["explorer"].fillna('na')

data.info()
```

```python
def convert_columns_to_datatypes (df, column_datatypes):
    for column, datatype in column_datatypes.items():
        if column in df.columns:
            df[column] = df[column].astype(datatype)
    return df

column_datatypes = {
'rank': int,
'supply' :float,
'maxSupply' :float,
'marketCapUsd' :float,
'volumeUsd24Hr': float,
'priceUsd':float,
'changePercent24Hr' :float,
'vwap24Hr': float
}

data = convert_columns_to_datatypes(data, column_datatypes)

data.info()
```

### Load

```python
db_username = "postgres"
db_password = getpass.getpass()
db_host = 'localhost'
db_port = 5432
db_name = 'CryptoStore'
```

```python
import psycopg2
from sqlalchemy import create_engine # manage and reuse database connections
```

```python
# Establish a connection
engine = create_engine(f'postgresql://{db_username}:{db_password}@{db_host}:{db_port}/{db_name}')

# table name
table_name = 'coincap_data'

# Load the data
data.to_sql(table_name, engine, if_exists='replace', index=False)

# Close connection
engine.dispose()
```
